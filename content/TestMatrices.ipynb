# Oefeningen: Matrices

Een matrix schrijven we in python als volgt:<br>
A = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
<br>Een matrix in python is dus een lijst van lijsten.
Zo is bijvoorbeeld <br>A[0][0] = 1

## Voorbeeldoefening: Eerste rij 
Opdracht: Vul de functie aan zodat deze de eerste rij van een gegeven matrix print.

```python

A = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

def PrintEersteRij(matrix):
  return matrix[0]

PrintEersteRij(A)

#Hieronder hoef je niet te kijken, dit om te kijken of je het juist hebt
assert(PrintEersteRij(A)==[1, 2, 3]), "Fout, je print niet de eerste rij"
```
